### read image, show image, save image, plot dot, plot line, plot rectangle, rgb2gray (<font color=#FF0000>opencv + pylab</font>)

```python
import cv2
from pylab import *
```

read image

```python
im = cv2.imread('../images/jpg/0420001.jpg')
type(im)
```

show image, plot dot, plot line

```python
imPylab = cv2.cvtColor(im, cv2.COLOR_BGR2RGB) # remember to change BGR(opencv) to RGB(pylab)
imshow(imPylab)
x = [10, 10, 40, 40]
y = [20, 50, 20, 50]
plot(x, y, 'r*')
plot(x[:2],y[:2])
title('dot and line')
axis('off')
show()
```

save image, plot rectangle, put text

```python
imgText = im.copy()
each = {}
each['x1'] = 30
each['y1'] = 30
each['x2'] = 40
each['y2'] = 40
each['txt'] = 'hello!'

cv2.rectangle(imgText, (each['x1'], each['y1']), (each['x2'], each['y2']), (0, 255, 0), 1)
cv2.putText(imgText, each['txt'], (each['x1'], each['y1']-5), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 255, 0))
imshow(imgText)
axis('off')
show()
cv2.imwrite('/home/lili/codes/testPython/images/result/test.jpg', imgText)
```

rgb2gray

```python
im_gray = cv2.cvtColor(im, cv2.COLOR_BGR2GRAY)

gray()
subplot(1,2,1)
imshow(im)
title('original image')
axis('off')
subplot(1,2,2)
imshow(im_gray)
title('gray image')
axis('off')
show()
```

### plot rectangle, put text(<font color=#FF0000>PIL, Chinese</font>!!!)

```python
from PIL import Image
from PIL import ImageDraw
from PIL import ImageFont
```

```python
each = {}
each['x1'] = 30
each['y1'] = 30
each['x2'] = 40
each['y2'] = 40
each['txt'] = '!'
box = [each['x1'], each['y1'], each['x2'], each['y2']]

imgRec = Image.fromarray(im)
draw = ImageDraw.Draw(imgRec)
ttfont = ImageFont.truetype('/usr/share/fonts/truetype/DroidSansFallbackFull.ttf',20) # download from the Internet
draw.rectangle(box, outline=(0, 255, 0))
draw.text((box[0], box[1]-18), unicode(each['txt'],'utf-8'), fill=(255,0,0), font = ttfont)
imshow(imgRec)
axis('off')
type(imgRec)
#imgRec.show() # open a window
```
